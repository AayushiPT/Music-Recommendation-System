```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Dataset.csv')
df.head(3)
```

```python
df['Genre']=df['Genre'].str.replace(r'(?<!^)(?=[A-Z])', ' ', regex=True)
df.head(3)
```

```python
df=df[df['Genre'] != 'Bollywood' ]
df.head(30)
```

```python
df.shape
```

```python
df['Genre'] = df['Genre'].str.replace('Bollywood', '', regex=False).str.strip()
df.head(10)
```

```python
df=df.sample(n=2200).reset_index(drop=True)
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfid = TfidfVectorizer(stop_words='english')
matrix = tfid.fit_transform(df['Genre']) 
print("TF-IDF Matrix Shape:", matrix.shape)
```

```python
similarity = cosine_similarity(matrix)
print("Similarity Matrix Shape:", similarity.shape)  # Verify dimensions

df[df['Genre'] == 'Dance']
```

```python
distance= sorted(list(enumerate(similarity[0])), reverse=False, key=lambda x:x[1])
```

```python
def recommendation(genre):
    if genre not in df['Genre'].unique():
        print(f"Genre '{genre}' not found in the dataset.")
        return []

    idx = df[df['Genre'] == genre].index[0]

    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])

    recommended_songs = []
    for i in distances[1:11]:  # Skip the first as it's the input song itself
        recommended_songs.append({
            "Song-Name": df.iloc[i[0]]['Song-Name'],
            "Singer/Artists": df.iloc[i[0]]['Singer/Artists'],
            "Genre": df.iloc[i[0]]['Genre'],
            "Album/Movie": df.iloc[i[0]]['Album/Movie'],
            "User-Rating": df.iloc[i[0]]['User-Rating']
        })

    return recommended_songs
```

```python
genre_input = "Romance"  # Replace with a genre present in your dataset

# Get recommendations
recommendations = recommendation(genre_input)

# Display recommendations
print(f"Top recommendations for genre '{genre_input}':")
for idx, song in enumerate(recommendations):
    print(f"{idx+1}. {song['Song-Name']} by -  {song['Singer/Artists']}\n  "
          f"\t \t(Genre: {song['Genre']}, Album: {song['Album/Movie']}, Rating: {song['User-Rating']})")
```

```python
import heapq  # Priority queue for A* implementation

def heuristic(current_index, target_index):
    return 1 - similarity[current_index][target_index]  # Smaller values = more similar

def a_star_recommendation(start_genre, target_genre, max_recommendations=10):
    
    # Filter songs by genre
    start_indices = df[df['Genre'] == start_genre].index.tolist()
    target_indices = df[df['Genre'] == target_genre].index.tolist()

    if not start_indices or not target_indices:
        print(f"Either '{start_genre}' or '{target_genre}' genre not found in the dataset.")
        return []

    # Priority queue for A* (score, current_index, path)
    pq = []
    for start_index in start_indices:
        heapq.heappush(pq, (0, start_index, [start_index]))

    visited = set()  # To avoid revisiting nodes
    recommendations = []

    while pq and len(recommendations) < max_recommendations:
        current_score, current_index, path = heapq.heappop(pq)

        if current_index in visited:
            continue
        visited.add(current_index)

        # Check if the current node is in the target genre
        if current_index in target_indices:
            recommendations.append({
                "Song-Name": df.iloc[current_index]['Song-Name'],
                "Singer/Artists": df.iloc[current_index]['Singer/Artists'],
                "Genre": df.iloc[current_index]['Genre'],
                "Album/Movie": df.iloc[current_index]['Album/Movie'],
                "User-Rating": df.iloc[current_index]['User-Rating']
            })

        # Expand neighbors
        for neighbor_index, similarity_score in enumerate(similarity[current_index]):
            if neighbor_index not in visited and neighbor_index != current_index:
                # Calculate A* score: cost (1 - similarity) + heuristic
                cost = 1 - similarity[current_index][neighbor_index]
                h = heuristic(neighbor_index, target_indices[0])  # Heuristic to first target node
                total_score = cost + h
                heapq.heappush(pq, (total_score, neighbor_index, path + [neighbor_index]))

    return recommendations
```

```python
# Input genres
start_genre = "Dance"
target_genre = "Romantic"  # Recommendations targeting this genre

# Get recommendations using A* algorithm
recommendations = a_star_recommendation(start_genre, target_genre)

# Display recommendations
print(f"Top recommendations from '{start_genre}' to '{target_genre}':")
for idx, song in enumerate(recommendations):
    print(f"{idx+1}. {song['Song-Name']}  by   {song['Singer/Artists']} "
          f"\n \t \t(Genre: {song['Genre']}, Album: {song['Album/Movie']}, Rating: {song['User-Rating']})")
```
